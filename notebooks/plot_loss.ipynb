```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
train = np.load("../checkpoints/train_loss_base.npy")
val = np.load("../checkpoints/val_loos_base.npy")

fig, ax = plt.subplots()
ax.set_yscale("log")
ax.plot(train)
ax.plot(val)
ax.set_ylabel("MSE")
ax.set_xlabel("training epoch (x50)")
ax.grid()
plt.draw()
plt.show()
```
