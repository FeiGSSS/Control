```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
train = np.load("../res/train_loss_batch.npy")
val = np.load("../res/val_loss_batch.npy")

fig, ax = plt.subplots()
ax.set_yscale("log")
ax.plot(train)
ax.plot(val)
ax.set_ylabel("MSE")
ax.set_xlabel("training epoch (x50)")
ax.grid()
plt.legend(["train_loss", "val_loss"])
#plt.savefig("../figs/loss_base_ode.pdf")
plt.draw()
plt.show()
print("Val Loss = {:.7f}".format(val[-1]))
```
